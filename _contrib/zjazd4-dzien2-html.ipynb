```python
URL = 'https://www.livescores.com/'
html = BeautifulSoup(requests.get(URL).text)
table = html.find(id='leagueTableBodyContainer')

result = []

def get(league):
    for row in league.find_all(attrs={'class': 'rows'}):
        yield {
            'rank': int(row.find(attrs={'data-type': 'rank'}).text),
            'name': row.find(attrs={'data-type': 'name'}).text,
            'goaldiff': int(row.find(attrs={'data-type': 'goaldiff'}).text),
            'points': int(row.find(attrs={'data-type': 'points'}).text)}

    # return [{'rank': int(row.find(attrs={'data-type': 'rank'}).text),
    #          'name': row.find(attrs={'data-type': 'name'}).text,
    #          'goaldiff': int(row.find(attrs={'data-type': 'goaldiff'}).text),
    #          'points': int(row.find(attrs={'data-type': 'points'}).text)}
    #         
    #          for row in league.find_all(attrs={'class': 'rows'})]
        
        
for league in html.find_all(attrs={'class': 'table'}):
    title = league.find(attrs={'data-type': 'league-name'}).attrs.get('title')
    
    if not title:
        continue
        
    country, name = title.split(' - ')
    result.append({
        'country': country,
        'name': name,
        'standing': pd.DataFrame(get(league)).set_index('rank'),
    })
    
    
result
```

```python
HEADERS = 

COMPETITION_NAMES = requests.get(
    url='https://cds-api.bwin.com/bettingoffer/counts?x-bwin-accessid=NTZiMjk3OGMtNjU5Mi00NjA5LWI2MWItZmU4MDRhN2QxZmEz&lang=en&country=PL&userCountry=PL&state=Latest&tagTypes=Sport,Region,Tournament,Competition,VirtualCompetition,VirtualCompetitionGroup&extendedTags=Sport,Region,Tournament,Competition,VirtualCompetition,VirtualCompetitionGroup&sportIds=4&sortBy=Tags&participantMapping=All',
    headers=HEADERS).json()

COMPETITION_STANDINGS = requests.get(
    url='https://cds-api.bwin.com/statistics/competition-batch?x-bwin-accessid=NTZiMjk3OGMtNjU5Mi00NjA5LWI2MWItZmU4MDRhN2QxZmEz&lang=en&country=PL&userCountry=PL&competitionIds=10330,18826,16152,55895,58583,56814&sportId=4',
    headers=HEADERS).json()

COLUMNS = {
    'team': 'Team',
    'matchesTotal': 'Played',
    'winTotal': 'Wins',
    'drawTotal': 'Draws',
    'lossTotal': 'Losses',
    'points': 'Points',
    'goalsTotal': 'Goals',
    'changeTotal': 'Change',
    'positionTotal': 'Position',
}


def get_league_name(competition_id):
    for row in COMPETITION_NAMES:
        competition = row['tag']
        if competition.get('parentLeagueId') == competition_id:
            return competition['name']['value']

        
result = {}

for standing in COMPETITION_STANDINGS:
    df = pd.DataFrame(standing['rows'])
    df['team'] = df['team'].apply(lambda x: x['teamName'])
    df.rename(columns=COLUMNS, inplace=True)
    df.set_index('Position', inplace=True)

    name = get_league_name(standing['competitionId'])
    result[name] = df

result
```
